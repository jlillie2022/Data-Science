#pie chart

```python
import matplotlib.pyplot as plt #creating a pie chart using data

labels=['frogs','dogs','cats','birds']
sizes=[15,30,45,10]
explode=(0,0.1,0,0)

figl, ax1= plt.subplots()
ax1.pie (sizes, explode=explode, labels=labels, autopct= '%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
```
